```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?

import seaborn as sns
import matplotlib.pyplot as plt

data = {'COL1':  [1, 2, 3, 4],
        'COL2': [5, 6, 7, 8],
        'COL3': [9, 10, 11, 12]}

df = pd.DataFrame(data)


corr1_Mat= df.corr() # calculate the coorelation matrix


sns.heatmap(corr1_Mat, annot=True, cmap='viridis', fmt=".2f")

plt.xlabel("Rows")
plt.ylabel("Columns")

plt.title("Seaborn & Heatmap")

plt.show()
```

```python
#5   Generate a bar plot using Plotly ?

import plotly.express as px
import pandas as pd


data2 = {'Years': ['2002', '2003', '2004', '2005'],
        'Profit': [15, 20, 12, 18]}
df = pd.DataFrame(data2)

# Generate a bar plot using Plotly Express
fig = px.bar(df, x='Years', y='Profit', title='Business Sample Plot')
fig.update_layout(
    width=400,
    height=300)


fig.show()
```

```python
#8  Create a line plot with multiple lines using Matplotlib ?

import matplotlib.pyplot as plt
import numpy as np


x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
y2 = np.array([1, 3, 5, 7, 9])


plt.step(x, y1, label='Line 1', marker='o')
plt.step(x, y2, label='Line 2', marker='x')


plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.legend()

plt.show()
```

```python
#9  Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold ?

import pandas as pd


data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 22, 35]}
df = pd.DataFrame(data)


threshold = 25


New_obj = df[df['Age'] > threshold]

print("\nFiltered DataFrame:")
print(New_obj)
```

```python
#10  Create a histogram using Seaborn to visualize a distribution ?

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


d1 = np.random.randn(1200)
data = pd.DataFrame(d1)
sns.histplot(data, kde=True)

plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#13  A Create a 3D scatter plot using Plotly ?

import plotly.express as px
import numpy as np
import pandas as pd


np.random.seed(42) # for reproducibility
num_points = 100
x_data = np.random.rand(num_points) * 10
y_data = np.random.rand(num_points) * 10
z_data = np.random.rand(num_points) * 10


data3 = {'X': x_data, 'Y': y_data, 'Z': z_data}
obj= pd.DataFrame(data3)

fig = px.scatter_3d(obj, x='X', y='Y', z='Z',
                    title='3D Scatter Plot')


fig.update_layout(margin=dict(l=0, r=0, b=0, t=40),width = 400, height = 500)


fig.show()
```
